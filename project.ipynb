```python
url="https://insights.blackcoffer.com/ai-and-its-impact-on-the-fashion-industry/"
```

```python
page=requests.get(url)
page.content
```

```python
soup = BeautifulSoup(page.content,'html.parser')
article = soup.findAll(attrs= {'class':'td-post-content tagdiv-type'})
text=article[0].text.replace('\n',"")
```

```python
head = soup.findAll('h1',attrs= {'class':'entry-title'})
Heading=head[0].text
```

```python
print(f"Heading\n{Heading}\nText\n{text}")
```
